```python
import requests
from bs4 import BeautifulSoup
import re

response = requests.get("https://www.thsrc.com.tw/tw/TimeTable/SearchResult")
print(response.encoding)
soup = BeautifulSoup(response.text, "lxml")
```

```python
soup.find_all("option", {"value":re.compile("[a-z0-9]{8}-[a-z0-9]{4}")})
```

```python
print(soup.find("section", class_ = "common_content"))
```

```python
form_data = {"StartStation":"2f940836-cedc-41ef-8e28-c2336ac8fe68",
             "EndStation":"e6e26e66-7dc1-458f-b2f3-71ce65fdc95f",
             "SearchDate":"2018/12/10",
             "SearchTime":"20:30",
             "SearchWay":"DepartureInMandarin"}
response_post = requests.get("https://www.thsrc.com.tw/tw/TimeTable/SearchResult", data = form_data)
soup_post = BeautifulSoup(response_post.text, "lxml")

soup_post.find("section", class_ = "common_content").find("tr")
```

```python
form_data = {"StartStation":"2f940836-cedc-41ef-8e28-c2336ac8fe68",
             "EndStation":"f2519629-5973-4d08-913b-479cce78a356",
             "SearchDate":"2018/12/10",
             "SearchTime":"07:30",
             "SearchWay":"DepartureInMandarin"}

# requests 改用 POST，並放入剛剛寫好的 dictionary
response_post = requests.get("https://www.thsrc.com.tw/tw/TimeTable/SearchResult", data = form_data)
soup_post = BeautifulSoup(response_post.text, "lxml")

# 觀察高鐵網頁後，發現車次資訊藏在 屬性 class=column1 的標籤裡面
print(len(soup_post.find_all("td", {"class":"column1"})))
```
